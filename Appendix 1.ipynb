## Conditional operations

```python
pd_df > 150
```

```python
pd_df['TSLA'] > 150
```

```python
pd_df.loc[pd_df["GOOG"]>580]
```

```python
pd_df.loc[(pd_df["GOOG"]>580) & (pd_df["TSLA"]>121)]
```

```python
pd_df.loc[~(pd_df["GOOG"]>580)]
```
